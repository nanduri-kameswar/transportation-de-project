```python
import pandas as pd

files = [
    {
        "file": "bulk_rides.json"
    },
    {
        "file": "map_cancellation_reasons.json"
    },
    {
        "file": "map_cities.json"
    },
    {
        "file": "map_payment_methods.json"
    },
    {
        "file": "map_ride_statuses.json"
    },
    {
        "file": "map_vehicle_makes.json"
    },
    {
        "file": "map_vehicle_types.json"
    }
]

for item in files:
    url = f"https://dltransportationproject.blob.core.windows.net/raw/ingestion/{item['file']}?{query_params}"

    df = pd.read_json(url)
    df_spark = spark.createDataFrame(df)

    # Writing data to the bronze layer
    df_spark.write.format("delta")\
                .mode("overwrite")\
                .option("overwriteSchema",  'true')\
                .saveAsTable(f"transportation.bronze.{item['file'].replace(".json", "")}")
```

```python

url = "https://dltransportationproject.blob.core.windows.net/raw/ingestion/bulk_rides.json?{query_params}"

df = pd.read_json(url)
df_spark = spark.createDataFrame(df)
if not spark.catalog.tableExists("uber.bronze.bulk_rides"):
    df_spark.write.format("delta")\
            .mode("overwrite")\
            .saveAsTable(f"uber.bronze.bulk_rides")
    print("This will not run more than 1 time")
```
